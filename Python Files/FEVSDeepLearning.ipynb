```python
logdir = os.path.join("logs", datetime.datetime.now().strftime("%Y%m%d-%H%M%S"))
tensorboard_callback = tf.keras.callbacks.TensorBoard(logdir, histogram_freq=1)
model1 = Sequential()
# 1st hidden layer
model1.add(Dense(30, activation='relu', input_dim=76))
# 2nd hidden layer
model1.add(Dense(20, activation='relu'))
model1.add(Dense(10, activation='relu'))
# Output layer
model1.add(Dense(1, activation='sigmoid'))
# Compilation with weighted metrics
adam1 = tf.keras.optimizers.Adam(learning_rate=0.001)
model1.compile(loss='binary_crossentropy', optimizer=adam1, metrics=['accuracy'])

model_output = model1.fit(X_train, y_train, epochs=4, batch_size=50, verbose=1,
                         validation_data=(X_test, y_test), shuffle=True, callbacks=[tensorboard_callback])
metric_list = model1.evaluate(X_test, y_test, batch_size=1, verbose=0)

print('keras_evaluate_accuracy=%.3f' %metric_list[1])
```
